```python
import requests
import json
import pandas as pd
from urllib import request, parse
```

```python
url = request.Request("https://api.bithumb.com/public/ticker/ALL_KRW")
response_data = request.urlopen(url)
res_data = response_data.read().decode('utf-8')
result = json.loads(res_data)
data = result["data"]
```

```python
basicTable = {'Symbol':[], "opening_price" :[], "closing_price":[],"min_price":[],"max_price":[],"fluctate_24H":[],"fluctate_rate_24H":[]}
coinTable = pd.DataFrame(data=basicTable)
coinTable
```

```python
keys_list = list(data)
keys_list
```

```python
for v in range(0, len(keys_list)-2):
    coin_data = {'Symbol':[keys_list[v]], "opening_price" :[result.get("data").get(keys_list[v]).get("opening_price")], "closing_price":[result.get("data").get(keys_list[v]).get("closing_price")],"min_price":[result.get("data").get(keys_list[v]).get("min_price")],"max_price":[result.get("data").get(keys_list[v]).get("max_price")],"fluctate_24H":[result.get("data").get(keys_list[v]).get("fluctate_24H")],"fluctate_rate_24H":[result.get("data").get(keys_list[v]).get("fluctate_rate_24H")]}
    print(coin_data)
    data_Frame = pd.DataFrame(data=coin_data)
    coinTable = coinTable.append(data_Frame, ignore_index=True)
```

```python
coinTable
```

```python
coinTable.to_csv("coinTotal.csv",index=False)
```
